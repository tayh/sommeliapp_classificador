```python
dls = ImageDataLoaders.from_folder('/content/gdrive/My Drive/Colab Notebooks/food_dataset', valid_pct=0.2, item_tfms=Resize(224))
dls.valid_ds.items[:3]
```

```python
learn = cnn_learner(dls, resnet34, metrics=[error_rate,accuracy])
```

```python
learn.export('/content/gdrive/My Drive/Colab Notebooks/classificador1.pkl')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(10,10))
```
